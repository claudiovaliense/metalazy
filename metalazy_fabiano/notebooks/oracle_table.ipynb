```python
path = '/home/lfmendes/data/mestrado/metalazy/results/oracle.csv'

results = {}
results['ORACLE'] = {}
results['NB'] = {}
results['LOGISTIC'] = {}
results['EXTRA\_TREES'] = {}

depara = {'oracle':'ORACLE',
          'nb':'NB',          
          'extraRf200': 'EXTRA\_TREES',
          'logistic': 'LOGISTIC'
         }

classifier = {}
f = open(path, "r")
current_dataset = ''
for x in f:    
    #print(x.replace('\n','').split(','))
    
    line = x.replace('\n','').split(',')
    print(line)
    if('dataset' in line[0]):
        current_dataset = line[0].replace('dataset_','')
        print(current_dataset)
        continue   
    clf = line[0].replace('all_features_','')         
    if clf in depara:
        results[depara[clf]][current_dataset] = {}
        results[depara[clf]][current_dataset]['macF1'] = "{0:.2f}".format(float(line[1])*100) 
        results[depara[clf]][current_dataset]['micF1'] = "{0:.2f}".format(float(line[2])*100) 
        results[depara[clf]][current_dataset]['macF1_std'] = "{0:.2f}".format(float(line[3])*100) 
        results[depara[clf]][current_dataset]['micF1_std'] = "{0:.2f}".format(float(line[4])*100) 
    else:
        print('not in depara: {}'.format(line[0]))
print(results)  
#print(results['BERT'])

print('\n\n')
initial = '\\begin{table*}[ht]\n\caption{Oracle MetaLazy Results}\n\label{tab:oracleresults}\n\\begin{tabular}{cccccccc}\n\\toprule\n&&20NG&REUT90&4UNI&ACM&STANFORD&YELP\\\\\n\midrule'
print(initial)
for clf in sorted(results.keys()):
    #print(clf)
    #print(results[clf])
    each1='\multirow{2}{8em}{'+ clf +'} &macF1&$'+ results[clf]['20NG']['macF1'] +'('+results[clf]['20NG']['macF1_std']+')$& $'+ results[clf]['REUT90']['macF1']+'('+results[clf]['REUT90']['macF1_std']+')$& $'+ results[clf]['4UNI']['macF1'] +'('+results[clf]['4UNI']['macF1_std']+')$&$'+ results[clf]['ACM']['macF1'] +'('+results[clf]['ACM']['macF1_std']+')$&$'+ results[clf]['STANFORD']['macF1'] +'('+results[clf]['STANFORD']['macF1_std']+')$&$'+ results[clf]['YELP']['macF1'] +'('+results[clf]['YELP']['macF1_std']+')$ \\\\'
    print(each1)
    
    each1='&micF1&$'+ results[clf]['20NG']['micF1'] +'('+results[clf]['20NG']['micF1_std']+')$& $'+ results[clf]['REUT90']['micF1'] +'('+results[clf]['REUT90']['micF1_std']+')$& $'+ results[clf]['4UNI']['micF1'] +'('+results[clf]['4UNI']['micF1_std']+')$& $'+ results[clf]['ACM']['micF1'] +'('+results[clf]['ACM']['micF1_std']+')$& $'+ results[clf]['STANFORD']['micF1'] +'('+results[clf]['STANFORD']['micF1_std']+')$& $'+ results[clf]['YELP']['micF1'] +'('+results[clf]['YELP']['micF1_std']+')$ \\\\'
    print(each1)
    
    print('\midrule')
    
final = '\\bottomrule\n\end{tabular}\n\end{table*}\n'
print(final)
```
